### （1）切片与范围关系

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
data1=np.random.randint(1,100,size=5)
data1
```

```python
data1[0]    #单一使用索引下标获得的只是一个值，不是一个数组对象
data1[[0,2]]    #通过list列表的索引下标读取的多个值，得到的是一个数组对象，并非一个值
```

```python
data1[2:4] #理解：其实切片只是使用索引一个范围区间的表示手法而已，语法为“起始下标号:结束下标号”【注意：这种方式得到的是一个数组对象，不是值】
```

```python
ds1=Series(data=data1)  #series对象的索引访问、切片的方法与np.array是一样的
# ds1[0]
# ds1[1:3]
# ds1[:3]
ds1[[1,3]]
```

```python
data2=np.random.randint(1,100,size=(4,5))
df2=DataFrame(data=data2)
df2
```

```python
df2[[0,3]] #读取多行
df2[1:4] #使用索引区间读取多行【注意：其实二维数组读取可以看作是两个一维数组组成，也就是用两个下标来表示[行][列]】
df2[::-1]
df2[::2] #指定步长间隔来读取多行
# df2.iloc[:,[1,3]] #【注意：如果是DataFrame对象的话，应该使用iloc的隐式索引（或者用创建DataFrame对象时用index指定的显式索引）切片才行】
# data2[:,[1,3]] #此方法只适合于数组，而对DataFrame对象是出错的（因为df2创建是没有用index指定显式索引为数字序号，但df2[:,[1,3]]默认会使用显式索引，但index又未指定，所以出错）
# df2.iloc[1:3,2:4]
```

```python
dict3={
    'python':99,
    'java':100,
    'C':88
}
dict3['java']
```

### 总结：
    1. numpy可以保持与python列表的访问方法，但是要注意高维数组的访问可能存在的潜在风险
    2. 多维数组直接访问元素：以列表的方式表示索引[d0_loc,d1_loc,...]
    3. 多维数组的行访问：n[row_index]，访问多列：n[[row_index1,row_index2,...]]
    4. 多维数组的列访问：n[col_index]，访问多列：n[[col_index1,col_index2,...]]
    5. BOOL列表访问：
        行访问：n[BOOLLIST]
        列访问：n[:,BOOLLIST]
        【注意】BOOLLIST的长度必须与访问的维度的长度一致

### (4)空值处理：python的空值是None；numpy的空值是np.nan(显示为NaN)  【注意：np.nan无论与谁运算结果都为nan】

```python
n1=np.array([1,2,3,np.nan])
n1.sum()  # 【注意：可以直接用调用对象属性的方法调用】
display(np.nansum(n1))  #【注意：使用numpy函数进行计算的调用方法，是用调用过程函数的方式进行<传参数>】
```

### (5)矩阵运算

```python
# 矩阵+ -法
n1=np.array([[1,2],[3,4]])
n2=np.array([[5,6],[7,8]])
display(n1,n2)
```

```python
# 矩阵加法+  【注意：都是两个数组之间对应位置的运算】
n1+n2
```

```python
# 矩阵减法-  【注意：都是两个数组之间对应位置的运算】
n1-n2
```

```python
# 矩阵乘法*  【注意：都是两个数组之间对应位置的运算】
n1*n2
```

```python
# 矩阵乘积  【注意：1、需要用np.dot()函数来运算；2、两个数组的乘积运算是按A数组的行中的每个数*B数组的列中的每个数，然后再求和得到一个数】
np.dot(n1,n2)
```

运算原理：
A_array([[1, 1],  
        [2, 2]])  
B_array([[3, 3],  
        [4, 4]])  

19=(A)1*(B)5+(A)2*(B)7       (A数组中的第1行对应数XB数组中的第1列对应数，然后把每个对应数相加所得的和即为结果数组的第1个值)  
22=(A)1*(B)6+(A)2*(B)8       (A数组中的第1行对应数XB数组中的第2列对应数，然后把每个对应数相加所得的和即为结果数组的第2个值)  
43=(A)3*(B)5+(A)4*(B)7       (A数组中的第2行对应数XB数组中的第1列对应数，然后把每个对应数相加所得的和即为结果数组的第3个值)  
50=(A)3*(B)6+(A)4*(B)8       (A数组中的第2行对应数XB数组中的第2列对应数，然后把每个对应数相加所得的和即为结果数组的第4个值)  

#### 矩阵运算--广播机制

```python
n3=np.random.randint(1,100,size=(3,3))
n4=np.random.randint(1,100,size=3)
display(n3,n4)
```

```python
n3+n4   #虽然两个数组的维度不一致，但列数一致；此时numpy会自动采用广播机制来匹配计算
```

#### 广播机制的自动匹配运算原理  
总原则是根据两个数组的行列维度数，来取行维度数和列维度数等中的最大数来重新组合成一个大数组；然后分别以此最大的行列维度数来拓展不够的哪个维度；两个数组经过拓展后会自动形成相同的维度数的新数组，最后再把拓展后的两个数组进行对应位置的运算即可
