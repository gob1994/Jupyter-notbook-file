## （1）pandas绘图

```python
from pandas import Series,DataFrame
import numpy as np
```

### 1、Series绘图

```python
data1=np.random.randn(10)
ls1=Series(data=data1)
```

```python
ls1.plot()
```

```python
#中文乱码的原因：pandas画图的内核是matplotlib或者说调用了它的api进行绘图，所以需要使用matplotlib才能解决中文显示乱码
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['Microsoft YaHei','SimHei']  #用来正常显示中文标签
# ls1.index=list('ABCDEFGHIJ')
ls1.index=['北京','上海','广州','深圳','南京','东莞','长春','成都','重庆','拉萨']
ls1.plot(color='g',linestyle='--',marker='.',figsize=[10,5],title='区域销售表')
```

```python
ls1.plot(color='r',kind='bar')
```

### 2、DataFrame绘图

```python
data2=np.random.randn(3,3)
df2=DataFrame(data=data2,columns=['python','java','php'],index=['eric','jojouly','alex'])
df2.plot()
```

```python
#中文乱码的原因：pandas画图的内核是matplotlib或者说调用了它的api进行绘图，所以需要使用matplotlib才能解决中文显示乱码
# import matplotlib.pyplot as plt
# plt.rcParams['font.sans-serif']=['Microsoft YaHei','SimHei']  #用来正常显示中文标签
df3=DataFrame(data=data2,columns=['语文','数学','English'],index=['张三','李四','王五'])
df3.plot()
```

## （2）matplotlib绘图

```python
#matplotlib图表中文乱码的解决方法
# import matplotlib.pyplot as plt
# plt.rcParams['font.sans-serif']=['Microsoft YaHei','SimHei']  #用来正常显示中文标签
```

```python
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt2
```

### 1、数组和Series作为数据源绘图

```python
data3=np.random.randn(10)
```

```python
plt2.plot(data3)    #matplotlib.plot允许直接使用数组来绘图
```

```python
plt2.plot(data3,color='g',linestyle='--',marker='.')    #证明pandas和matplotlib里plot属性和方法是一致的；其实pandas绘图核心使用的就是matplotlib的绘图核心，所以他们是相同的属性和方法
```

```python
ls2=Series(data=data3)
# ls2.index=list('ABCDEFGHIJ')
ls2.index=['北京','上海','广州','深圳','南京','东莞','长春','成都','重庆','拉萨']
plt2.plot(ls2)  #允许使用Series对象的数据
```

### 2、DataFrame作为数据源绘图

```python
plt2.plot(df4)
```

```python
df5=DataFrame(np.random.randint(1,100,size=(10,4)).cumsum(1),index=['北京','上海','广州','深圳','南京','东莞','长春','成都','重庆','拉萨'],columns=['python','java','c','php'])
df5
```

```python
plt2.plot(df5)
```

```python
fig3,axes3=plt.subplots(1,2,figsize=[10,8])
```

```python
x=np.arange(0,5,0.2)
y1=np.sin(x)
y2=np.e**(0.4*x)-2
y3=np.cos(x)
y4=0.5*x
```

```python
axes3[0].plot(x,y1,label='sin')
axes3[0].plot(x,y2,label='exp1')
fig3
```

```python
axes3[1].plot(x,y3,'--',linewidth=2.7,label='cos')
axes3[1].plot(x,y4,'-.',label='linear')
fig3
```
